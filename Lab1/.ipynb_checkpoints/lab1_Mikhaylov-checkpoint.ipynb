Подключение модулей

```python
import pandas                                          #библиотека для работы с данными, мы считываем с помощью него данные из файла
import tensorflow as tf                                #библиотека используется для обучения модели нейросети
import matplotlib.pyplot as plt                        #библиотека для построения графиков
from tensorflow.keras.layers import Dense              #для создания слоя
from tensorflow.keras.models import Sequential         #инициализация модели данных
from tensorflow.keras.utils import to_categorical      #для преобразования формата данных
from sklearn.preprocessing import LabelEncoder         #библиотека для подготовки данных и оценки модели.
```

Загрузка данных

```python
dataframe = pandas.read_csv("iris.csv", header=None) #считываем таблицу из в файла в переменную датафрейм
dataset = dataframe.values                           #преобразовываем в список 
X = dataset[:,0:4].astype(float)                     #делим значения на 2 категории (X - входные данные, Y - выходные данные)
Y = dataset[:,4]
```

Переход от текстовых меток к категориальному вектору

```python
encoder = LabelEncoder()           #инициализируем кодировщик данных
encoder.fit(Y)                     #устанавливает кодировщик
encoded_Y = encoder.transform(Y)   #преобразовывает к числовому формату
dummy_y = to_categorical(encoded_Y) #преобразование выходных атрибутов из вектора в матрицу 
```

Задаем базовую архитектуру сети

```python
model = Sequential()             #инициализирует модель данных
model.add(Dense(4, activation='relu'))  #создание слоя (relu - избавляемся от отрицательных чисел)
model.add(Dense(3, activation='softmax')) #создание слоя (softmax - преобразовываем в вероятность [0-1] )
```

Инициализация параметров обучения

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])  #указываем параметры обучения(optimizer - тип процедуры обучения, loss -функция потерь, metrics - имя или обьект из модуля metrics)
```

Обучение сети

```python
h = model.fit(X, dummy_y, epochs=75, batch_size=10, validation_split=0.1) #запускает обучение по модели (epochs - кол-во итераций, batch_size - кол-во выборов для одного обучения, validation_split - часть данных об обучении которые будут использоваться в качестве данных для проверки)
```

```python
plt.plot(h.history['loss'])
plt.plot(h.history['val_loss'])
plt.title('model loss')
plt.legend(['loss', 'val_loss'], loc='upper left')
plt.show()

plt.plot(h.history['accuracy'])
plt.plot(h.history['val_accuracy'])
plt.title('model accuracy')
plt.legend(['accuracy', 'val_accuracy'], loc='upper left')
plt.show()
```
